```python
import pandas as pd
from nltk.tokenize import word_tokenize
```

```python
data = pd.read_csv('tweets_cleaned_technology.csv')
```

```python
data
```

```python
document_id_name_decionary={}
docid_tokens={}
inverted_index={}
```

```python
for i in range(len(data)):
    document_id_name_decionary[i+1] = data.iloc[i]['id']
    
    text = word_tokenize(data.iloc[i]['text_cleaned'])
    
    docid_tokens[i+1]=text
    
    text=set(text)
    
    for word in text:
        if(word in inverted_index):
            inverted_index[word].append(i+1)
        else:
            inverted_index[word]=[i+1]
```

```python
filehandler = open("document_id_name_decionary.txt", 'wt',errors="ignore",encoding ="utf-8")
data = str(document_id_name_decionary)
filehandler.write(data)

filehandler = open("docid_tokens.txt", 'wt',errors="ignore",encoding ="utf-8")
data = str(docid_tokens)
filehandler.write(data)

filehandler = open("inverted_index.txt", 'wt',errors="ignore",encoding ="utf-8")
data = str(inverted_index)
filehandler.write(data)

filehandler.close()
```
